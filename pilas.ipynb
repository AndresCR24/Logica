8. E8. Cree una función que reciba una cola de colas. En estas colas hay objetos de tipo cliente con un atributo nombre y otro prioridad (1, 2 o 3). Esta función se enfocará en atender cada una de las colas internas por prioridad, es decir, primero atenderá a todos los de prioridad 1 de todas las colas, luego los de prioridad 2, etc. Cada vez que se atienda un cliente, saldrá su nombre, prioridad y la cola donde está.

```python
import copy

class EmptyStackException(Exception):
    pass

class Queue:
    def __init__(self):  # Corregido de __int__ a __init__
        self.queue = []

    def enqueue(self, item):  # Corregido la indentación
        self.queue.append(item)

    def dequeue(self):  # Corregido la indentación
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)

    def first(self):  # Corregido la indentación
        if len(self.queue) < 1:
            return None
        return self.queue[0]

    def __len__(self):  # Corregido la indentación
        return len(self.queue)

    def __repr__(self):  # Método agregado para representar el objeto como una cadena
        return str(self.queue)
    
    def guardar_estado(self):
            # Usamos copy.deepcopy para asegurar una copia profunda del estado de la pila
        self.estado_guardado = copy.deepcopy(self.queue)
            
    def restaurar_estado(self):
        # Restauramos usando la copia profunda guardada
        self.queue = copy.deepcopy(self.estado_guardado)
        
class Cliente:
    def __init__(self, nombre, prioridad):
        self.nombre = nombre
        self.prioridad = prioridad

    def __repr__(self):
        return f"Cliente({self.nombre}, Prioridad {self.prioridad})"


def atender_clientes_por_prioridad(cola_de_colas):
    # Se supone que hay 3 niveles de prioridad: 1, 2 y 3.
    for prioridad_actual in range(1, 4):  # Itera sobre cada nivel de prioridad
        print(f"\nAtendiendo clientes con Prioridad {prioridad_actual}:")
        
        cola_de_colas.guardar_estado()  # Guarda el estado inicial de la cola de colas
        
        for _ in range(len(cola_de_colas)):
            cola_actual = cola_de_colas.dequeue()
            cola_actual.guardar_estado()  # Guarda el estado de la cola actual

            nueva_cola = Queue()  # Crea una nueva cola para almacenar los clientes no atendidos aún
            
            while len(cola_actual) > 0:
                cliente_actual = cola_actual.dequeue()
                if cliente_actual.prioridad == prioridad_actual:
                    # Atiende al cliente
                    print(f"Atendiendo a {cliente_actual.nombre} de Prioridad {cliente_actual.prioridad}")
                else:
                    # Agrega el cliente a la nueva cola si no coincide la prioridad
                    nueva_cola.enqueue(cliente_actual)

            cola_de_colas.enqueue(nueva_cola)  # Encola la nueva cola con los clientes restantes
        cola_actual.restaurar_estado()
        cola_de_colas.restaurar_estado()  # Restaura el estado inicial de la cola de colas después de cada nivel de prioridad

# Crear clientes
cliente1 = Cliente("Ana", 1)
cliente2 = Cliente("Luis", 2)
cliente3 = Cliente("Sofía", 3)
cliente4 = Cliente("Carlos", 1)
cliente5 = Cliente("Betty", 3)

# Crear colas internas y agregar clientes
cola1 = Queue()
cola1.enqueue(cliente1)
cola1.enqueue(cliente2)

cola2 = Queue()
cola2.enqueue(cliente3)
cola2.enqueue(cliente4)

cola3 = Queue()
cola3.enqueue(cliente5)

# Crear la cola de colas y agregar las colas internas
cola_de_colas = Queue()
cola_de_colas.enqueue(cola1)
cola_de_colas.enqueue(cola2)
cola_de_colas.enqueue(cola3)
print(cola_de_colas)
# Atender a los clientes por prioridad
atender_clientes_por_prioridad(cola_de_colas)
print(cola_de_colas)
```

```python
import copy

class EmptyStackException(Exception):
    pass

class Stack:
    def __init__(self):
        self.stack = []
        self.estado_guardado = []

    def __len__(self):
        return len(self.stack)

    def push(self, dato):
        self.stack.append(dato)

    def pop(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack.pop()

    def top(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack[-1]

    def __repr__(self):
        return str(self.stack)
    
    def guardar_estado(self):
        self.estado_guardado = copy.deepcopy(self.stack)

    def restaurar_estado(self):
        self.stack = copy.deepcopy(self.estado_guardado)
        
        
def encontrar_pares_con_suma(s, n):
    # Usamos un conjunto para guardar los complementos de los números buscados
    complementos = set()
    
    # Guardamos el estado actual de la pila para restaurarlo al final
    s.guardar_estado()
    
    while len(s) > 0:
        actual = s.pop()
        # Si el número actual está en el conjunto de complementos, hemos encontrado los dos números
        if actual in complementos:
            s.restaurar_estado()  # Restauramos el estado original de la pila antes de retornar
            return True
        # Agregamos el complemento del número actual al conjunto
        complementos.add(n - actual)
    
    s.restaurar_estado()  # Restauramos el estado original de la pila si no encontramos pares
    return False

# Ejemplo de uso
s = Stack()
s.push(1)
s.push(4)
s.push(5)
s.push(6)

# Caso de prueba
N = 9
print(s)
print(encontrar_pares_con_suma(s, N))  # Debería retornar True si hay números que sumen 9, por ejemplo 4 y 5.
print(s)
```

10. E10. Escriba una función que reciba un string de paréntesis y diga si está balanceado. Este string estará balanceado si cada paréntesis que abre, tiene uno que cierra.
Por ejemplo, la expresión "()()()" es válida. La expresión "(()" es inválida La expresión "(()())()(())(((())))" es válida.

```python
import copy

class EmptyStackException(Exception):
    pass

class Stack:
    def __init__(self):
        self.stack = []
        self.estado_guardado = []

    def __len__(self):
        return len(self.stack)

    def push(self, dato):
        self.stack.append(dato)

    def pop(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack.pop()

    def top(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack[-1]

    def __repr__(self):
        return str(self.stack)
    
    def guardar_estado(self):
        self.estado_guardado = copy.deepcopy(self.stack)

    def restaurar_estado(self):
        self.stack = copy.deepcopy(self.estado_guardado)
        
def es_balanceado(expresion):
    pila = Stack()  # Utilizamos la pila para llevar el seguimiento de los paréntesis que abren
    
    for caracter in expresion:
        if caracter == '(':  # Si encontramos un paréntesis que abre, lo empujamos a la pila
            pila.push(caracter)
        elif caracter == ')':  # Si encontramos un paréntesis que cierra
            if len(pila) == 0:
                # No hay un paréntesis que abre correspondiente, por lo tanto no está balanceado
                return False
            else:
                pila.pop()  # Sacamos el paréntesis que abre correspondiente de la pila
    
    # Después de recorrer toda la expresión, si la pila está vacía, entonces está balanceado
    return len(pila) == 0

# Ejemplos de uso
print(es_balanceado("()()()"))  # Debería imprimir True
print(es_balanceado("(()"))  # Debería imprimir False
print(es_balanceado("(()())()(())(((())))"))  # Debería imprimir True
```

```python
import copy

class EmptyStackException(Exception):
    pass

class Stack:
    def __init__(self):
        self.stack = []
        self.estado_guardado = []

    def __len__(self):
        return len(self.stack)

    def push(self, dato):
        self.stack.append(dato)

    def pop(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack.pop()

    def top(self):
        if len(self.stack) == 0:
            raise EmptyStackException("La pila está vacía")
        return self.stack[-1]

    def __repr__(self):
        return str(self.stack)
    
    def guardar_estado(self):
        self.estado_guardado = copy.deepcopy(self.stack)

    def restaurar_estado(self):
        self.stack = copy.deepcopy(self.estado_guardado)
        
def es_balanceado(expresion, pila):
    # Asegurarse de que la pila esté vacía antes de comenzar
    while len(pila) > 0:
        pila.pop()

    for caracter in expresion:
        if caracter == '(':
            pila.push(caracter)
        elif caracter == ')':
            if len(pila) == 0:
                return False
            else:
                pila.pop()

    return len(pila) == 0

# Creación de la instancia de Stack fuera de la función
pila_para_uso_general = Stack()

# Ejemplos de uso
print(es_balanceado("()()()", pila_para_uso_general))  # Debería imprimir True
print(es_balanceado("(()", pila_para_uso_general))  # Debería imprimir False
print(es_balanceado("(()())()(())(((())))", pila_para_uso_general))  # Debería imprimir True

```
